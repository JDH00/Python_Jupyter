# 신경망 기초 수학

```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['seaborn-whitegrid'])
```

# 일차함수
- $y=ax+b$
    - $a$:기울기 $b$:$y$절편
- 그래프 상에서 직선인 그래프(linear)

```python
def linear_function(x):
    a=0.5
    b=2
    
    return a*x+b
```

```python
print(linear_function(5))
```

```python
x = np.arange(-5,5,0.1)
y=linear_function(x)

plt.plot(x,y) #plt.plot(y)만 하면 x범위가 0부터 나옴
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
plt.show()
```

# 이차함수
- $y=ax^2+bx+c$
- 일반적으로 두 개의 실근을 가짐

```python
def quadratic_function(x):
    a=1
    b=-1
    c=-2
    
    return a*x**2+b*x+c
```

```python
print(quadratic_function(2))
```

```python
x = np.arange(-5,5,0.1)
y=quadratic_function(x)

plt.plot(x,y) #plt.plot(y)만 하면 x범위가 0부터 나옴
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function')
plt.show()
```

# 삼차함수
- $y=ax^3+bx^2+cx+d$
- 일반적으로 두 개의 실근을 가짐

```python
def cubic_function(x):
    a=4
    b=0
    c=-1
    d=-8
    
    return a*x**3 + b*x**2 + c*x + d
```

```python
print(cubic_function(3))
```

```python
x = np.arange(-5,5,0.1)
y=cubic_function(x)

plt.plot(x,y) #plt.plot(y)만 하면 x범위가 0부터 나옴
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Function')
plt.show()
```

# 함수의 최솟값/최댓값

```python
def my_func(x):
    a=1
    b=-3
    c=10
    return a*x**2 + b*x + c
```

```python
x=np.arange(-10,10,0.01)
y=my_func(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(1.5, my_func(1.5))
plt.text(1.5-1.5, my_func(1.5)+10, 'min value of f(x)\n({},{})'.format(1.5, my_func(1.5)),fontsize=10)
plt.title('my_func')
plt.show()
```

# 특정 구간 내에서 최솟값 구하기

```python
def get_minimum(x1,x2,f):
    x=np.arange(x1,x2,0.01)
    y=f(x)
    
    plt.plot(x,y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Get Minimum')
    plt.show()
    
    return min(y)
```

```python
print(get_minimum(1,4,my_func))
```

# 지수함수/로그함수
- 지수함수-로그함수는 역함수 관계($y=x$ 직선 대칭 단, 밑이 같을 때)
- 파이썬으로 직접 구현 가능

# 지수함수
- $y=a^x (a\neq0) (기본형)$
- $y=e^x (e=2.71828...)$

```python
def exponential_function(x):
    a=4
    return a**x
```

```python
print(exponential_function(4))
print(exponential_function(0))
```

```python
x=np.arange(-3,2,0.1)
y=exponential_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,15)
plt.xlim(-4,3)
plt.title('Exponential Function')
plt.show()
```

```python
def exponential_function2(x):
    a=4
    return math.pow(a,x)
```

```python
print(exponential_function2(4))
print(exponential_function2(0))
```

# 밑이$e$인 지수함수

```python
print(math.exp(4))
print(np.exp(4))
```

# 로그함수
- $y=log_{a}{(x)} (a\neq1)(기본형)$
- $log_{10}{x}$ $(상용로그)$
- $y=ln(x) (밑이 e인 자연로그)$

```python
print(math.log(2,3))#밑이3 진수가2
print(np.log2(4))#밑이2인 로그함수
print(np.log(4))#밑이e인 자연로그
```

# 역함수 관계
- $y=x$

```python
x=np.arange(-1,5,0.01)
y1=np.exp(x)

x2=np.arange(0.00001,5,0.01)
y2=np.log(x2)

y3=x

plt.plot(x,y1,'r-', x2,y2,'b-', x,y3,'k--')

plt.ylim(-2,6)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# 함수 조작
- $y=-log{_a}{(x)}와 y=-log{_a}{(1-x)}$
    - $x=0.5 대칭$
- $Logistic Regression 을 위한 함수$

```python
x=np.arange(-10,10,0.01)
y1=-np.log(x)
y2=-np.log(1-x)

plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')

plt.grid()
plt.plot(x,y1,'b-',x,y2,'r-')
plt.text(0.9, 2.0, 'y=-log(1-x)',fontsize=15)
plt.text(0.1, 3.0, 'y=-log(x)',fontsize=15)
plt.xlim(-0.3,1.4)
plt.ylim(-0.5,4)
plt.scatter(0.5, -np.log(0.5))
plt.show()
```

# 극한
- 극한에 대해서는 어떠한 식을 코드로 표현할 수 있다 정도로만 이해하며 참고
- 극한에서 알아야 할 사실은 $x$가 어떤 값 $a$에 가까이 다가갈 때 a에 '한없이 가까이 간다'일 뿐, a에 도달하지는 않는다
- 이를 표현할 때, 엡실론(epsilon)이라는 아주 작은 값(ex. 0.0001)등으로 표현

```python
from sympy import *
init_printing()

x,y,z = symbols('x y z')
a,b,c,t = symbols('a b c t')
```

$lim_{x\to3}(\frac{x^3-1}{x-1})$

```python
print('극한값:', limit((x**3-1) /(x-1),x,1 ))

plot( ((x**3-1)/(x-1)),xlim=(-5,5),ylim=(-1,10) );
```

$lim_{x\to\infty}(\frac{1+x}{x})$

```python
print('극한값:', limit((1+x) /x,x,oo ))

plot((1+x)/x,xlim=(-10,10),ylim=(-5,5) );
```

$lim_{x\to\infty}(\frac{\sqrt{x+3}-2 }{x-1})=\frac{1}{4}$

```python
print('극한값:', limit( (sqrt(x+3)-2) / (x-1), x ,1))

plot((sqrt(x+3)-2)/(x-1),xlim=(-5, 12), ylim=(-0.5,1));
```
