```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
y=np.array([20,10,30,50,15])
plt.plot(y)
```

```python
x=np.array([10,30,15])
y=np.array([10,20,15])
plt.plot(x,y,'r-o',label='Data 1')
plt.ylabel('y Axis')
plt.xlabel('x Axis')
plt.legend()
```

```python
x=np.linspace(0,2,100)
y1=0.5*x
y2=0.5*x**2
y3=0.5*x**3
```

```python
plt.plot(x,y1,label='linear')
plt.plot(x,y2,label='quadratic')
plt.plot(x,y3,label='cubic')
plt.title('Graphs')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```
