# Plot Example 2 - Object-Oriented Style

```python
x
```

```python
y
```

```python
hf = plt.figure() # figure handle
ha = hf.add_axes([0,0,1,1]) # Axes handle
ha.plot(x,y) #axis handle
```

```python
fig, ax = plt.subplots()
line, = ax.plot(x,y)
```

```python
x_new = np.array([10,11,12])
y_new = np.array([10,10,10])
line.set_data(x_new,y_new)
fig
```

```python
i=1.1
x_new = np.array([10,11,12])
y_new = np.array([10,10,10])
line.set_data(x_new,i*y_new)
fig
# figure 와 handle을 쓰면 반복문을 이용해서 애니메이션으로 만들 수 있음
```
