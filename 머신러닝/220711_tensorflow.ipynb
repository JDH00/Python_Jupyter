# Machine learning

```python
# 라이브러리 사용
import pandas as pd
import tensorflow as tf
```

```python
# 데이터 준비
# 파일들로부터 데이터 읽어오기
파일경로1 ='https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv'
레몬에이드= pd.read_csv(파일경로1)

파일경로2 ='https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
보스톤= pd.read_csv(파일경로2)

파일경로3 ='https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
아이리스= pd.read_csv(파일경로3)
```

```python
# 데이터 모양으로 확인하기
print(레몬에이드.shape)
print(보스톤.shape)
print(아이리스.shape)
```

```python
# 칼럼이름 출력
print(레몬에이드.columns)
print(보스톤.columns)
print(아이리스.columns)
```

```python
독립1=레몬에이드[['온도']]
종속1=레몬에이드[['판매량']]
print(독립1.shape,종속1.shape)

독립2=보스톤[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat']]
종속2=보스톤[['medv']]
print(독립2.shape,종속2.shape)

독립3=아이리스[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
종속3=아이리스[['품종']]
print(독립3.shape,종속3.shape)
```

```python
레몬에이드.head()
```

```python
 보스톤.head()
```

```python
아이리스.head()
```

```python
# 모델을 만든다.
X= tf.keras.layers.Input(shape=[1])
Y=tf.keras.layers.Dense(1)(X)
model=tf.keras.models.Model(X,Y)
model.compile(loss='mse')
```

```python
# 모델을 학습(fit)시킨다.
model.fit(독립1,종속1,epochs=100)
```

```python
# 모델을 이용한다.
model.predict(독립1)
```

```python
model.predict([[15]])
```

```python
#원핫 인코딩
인코딩=pd.get_dummies(아이리스)
인코딩
```

```python
인코딩.columns
```

```python
독립3=인코딩[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
종속3=인코딩[['품종_setosa','품종_versicolor','품종_virginica']]
print(독립3.shape,종속3.shape)
```

```python
X3=tf.keras.layers.Input(shape=[4])
Y3=tf.keras.layers.Dense(3, activation='softmax')(X3)
model=tf.keras.models.Model(X3,Y3)
model.compile(loss='categorical_crossentropy',
             metrics='accuracy')
```

```python
model.fit(독립3,종속3,epochs=100)
```

```python
model.predict(독립3[-5:])
```

```python
종속3[-5:]
```

```python
#학습한 가중치
model.get_weights()
```

# 히든레이어, 여러개의 뉴런

```python
print(독립2.shape,독립2.shape)
```

```python
X2=tf.keras.layers.Input(shape=[13])
H2=tf.keras.layers.Dense(10, activation='swish')(X2)
Y2=tf.keras.layers.Dense(1)(H2)
model=tf.keras.models.Model(X2,Y2)
model.compile(loss='mse')
```

```python
model.summary()
```

```python
model.fit(독립2,종속2,epochs=1000,verbose=0)
model.fit(독립2,종속2,epochs=10)
```

```python
print(model.predict(독립2[:5]))
print(종속2[:5])
```

```python
X3=tf.keras.layers.Input(shape=[4])
H3=tf.keras.layers.Dense(8, activation='swish')(X3)
H3=tf.keras.layers.Dense(8, activation='swish')(H3)
H3=tf.keras.layers.Dense(8, activation='swish')(H3)
Y3=tf.keras.layers.Dense(3, activation='softmax')(H3)
model=tf.keras.models.Model(X3,Y3)
model.compile(loss='categorical_crossentropy',
             metrics='accuracy')
```

```python
model.fit(독립3,종속3,epochs=100)
```

```python
print(model.predict(독립3[0:5]))
```
